# **Kmeans**

```
from sklearn.cluster import KMeans
from sklearn.preprocessing import OneHotEncoder

# Assuming you have a DataFrame named 'df' with the features you want to cluster

# Select the features for clustering
features = train_df[['Automotive',
 'Baby Products',
 'Health & Beauty',
 'Media',
 'Consumer Electronics',
 'Console & Video Games',
 'Tools',
 'living',
 'Grocery',
 'Home',
 'Betting',
 'Jewellery & Watches',
 'Musical Instruments',
 'Office',
 'Pet Supplies',
 'Computer Software',
 'Sports',
 'Toys & Games',
 'Dating',
 'Clothing & Shoes'
#  'Age',
#  'Gender',
#  'Type of Job',
#  'Weekly working hours',
#  'Income',
#  'Home country'
 ]]

# Perform one-hot encoding on the string features
encoder = OneHotEncoder()
encoded_features = encoder.fit_transform(features)

# Perform K-means clustering for a range of cluster numbers
max_clusters = 10
inertias = []
for k in range(1, max_clusters+1):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(encoded_features)
    inertias.append(kmeans.inertia_)

# Plot the inertia values against the number of clusters
plt.plot(range(1, max_clusters+1), inertias, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Point of K-means Clustering')
plt.show()
```

# **DBSCAN**

```
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Assuming you have a DataFrame named 'df' with the one-hot encoded dataset

# Select the one-hot encoded features for clustering
featurevalue = features.values

# Standardize the features
scaler = StandardScaler()
featurevalue = scaler.fit_transform(featurevalue)

# Perform DBSCAN clustering
eps = 0.1  # Epsilon parameter
min_samples = 2  # Minimum number of samples in a cluster
dbscan = DBSCAN(eps=eps, min_samples=min_samples)
clusters = dbscan.fit_predict(features)

# Add the cluster labels back to the DataFrame
train_df['Cluster'] = clusters

# Print the cluster labels for each data point
print("Cluster Labels:")
print(train_df['Cluster'])
```

# **Linear Regression**

```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Assuming your dataframe is named 'df'
features = ['Feature1',
 'Feature2',
 'Feature3',
 'Feature4',
 'Feature5',
 'Feature6',
 'Feature7',
 'Feature8',
 'Feature9',
 'Feature10',
 'Feature11',
 'Feature12',
 'Feature13',
 'Feature14',
 'Feature15',
 'Feature16',
 'Feature17',
 'Feature18',
 'Feature19',
 'Feature20']
inputs = ['Openness to Experience', 'Conscientiousness', 'Extraversion', 'Agreeableness', 'Neuroticism']

for feature in features:
    # Extract the feature values and the corresponding inputs
    train_X = train_df[inputs].values
    train_y = train_df[feature].values
    val_X = val_df[inputs].values
    val_y = val_df[feature].values
    
    # Create and train the linear regression model
    model = LinearRegression()
    model.fit(train_X, train_y)
    
    # Make predictions using the trained model
    predictions = model.predict(val_X)
    
    # Print information about the model
    print(f"Feature: {feature}")
    print(f"Intercept: {model.intercept_}")
    print(f"Coefficients: {model.coef_}")
    print(f"Mean Squared Error: {mean_squared_error(val_y, predictions)}")
    print("")
```

# **Ridge Regression**

```
from sklearn.linear_model import Ridge
for feature in features:
    # Extract the feature values and the corresponding inputs
    train_X = train_df[inputs].values
    train_y = train_df[feature].values
    val_X = val_df[inputs].values
    val_y = val_df[feature].values
    
    # Create and train the linear regression model
    model = Ridge(alpha=1.0)
    model.fit(train_X, train_y)
    
    # Make predictions using the trained model
    predictions = model.predict(val_X)
    
    # Print information about the model
    print(f"Feature: {feature}")
    print(f"Intercept: {model.intercept_}")
    print(f"Coefficients: {model.coef_}")
    print(f"Mean Squared Error: {mean_squared_error(val_y, predictions)}")
    print("")
```
